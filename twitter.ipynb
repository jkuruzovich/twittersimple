```python
def get_hashtags(twitter, cf_dict, all_hashtags):
    dt = datetime.datetime.now()
    fn = dt.strftime('%Y-%m-%d-user-profiles.json')
    with open(fn, 'w') as f:
        # Create a subquery, looking up information about users listed in 'string'
        # twitter api-docs: https://developer.twitter.com/en/docs/accounts-and-users/follow-search-get-users/api-reference/get-users-lookup
        for hashtag in all_hashtags:
            try:
                out = twitter.search.tweets(q=hashtag)
                print(out)
                #out = "twitter.users.lookup(screen_name = name_set)"
                sub_start_time = time.time()
                  # Save user info
                f.write(json.dumps(out))
                f.write("\n")
            except TwitterHTTPError:
                print('\n-----------------\n')
                traceback.print_exc()
                time.sleep(cf_dict['sleep_interval'])
    return fn
output= get_hashtags(twitter, cf_dict, ["#iloveny", "#rpi"])
output
```
